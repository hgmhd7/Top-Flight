```python
# Import dependencies
import pandas as pd
import io
import requests
from config import gkey
import json
```

```python
# Get airport information from open flights endpoint
url="https://raw.githubusercontent.com/jpatokal/openflights/master/data/airports.dat"
request_url = requests.get(url).content
all_airports = pd.read_csv(io.StringIO(request_url.decode('utf-8')), header=None)
```

```python
# Add airport names to list for API calls
airport_name_list = []

for airport_name in all_airports["name"]:
    
    airport_name_list.append(airport_name)
    
len(airport_name_list)  
```

```python
# Tester for my target airport
target_airport = "st. louis lambert airport"

# Build the endpoint URL
target_url = ('https://maps.googleapis.com/maps/api/geocode/json?'
    'address={0}&key={1}').format(target_airport, gkey) 

# Run a request to endpoint and convert result to json
airport_geo_data = requests.get(target_url).json()

test_place_id = airport_geo_data["results"][0]["place_id"]

# # Print the json (pretty printed)
print(json.dumps(airport_geo_data, indent=4, sort_keys=True))
```

```python
# Check ID retrieval
test_place_id
```

# Retrieve API data ()

```python
# Set counter for API calls
counter = 0

# Set empty lists to store values
airport_google_name_list = []
airport_place_id_list = []
airport_full_address_list = []


# loop through airport names to retrive the Google names, place IDs, and addresses
for airport in airport_name_list:
    
    try:
       # Tester for my target airport
        target_airport = airport

        # Build the endpoint URL
        target_url = ('https://maps.googleapis.com/maps/api/geocode/json?'
            'address={0}&key={1}').format(target_airport, gkey) 

        # Run a request to endpoint and convert result to json
        airport_geo_data = requests.get(target_url).json()
        
        
        airport_name = airport_geo_data["results"][0]['address_components'][0]['long_name']
        airport_place_id = airport_geo_data["results"][0]["place_id"]
        airport_full_address = airport_geo_data["results"][0]["formatted_address"]
       
        
        
        airport_google_name_list.append(airport_name) 
        airport_place_id_list.append(airport_place_id)
        airport_full_address_list.append(airport_full_address)
       
        
        
        counter += 1 
            
            
        print(f'\nRetriveing data for {airport_name}.  \nAirport retrieval number {counter} complete!!!')
        
        
    
    except IndexError:
        
        airport_google_name_list.append('not found')
        airport_place_id_list.append('not found')
        airport_full_address_list.append('not found')
       
    
        counter += 1
        print(f'\n ERROR retriveing airport number {counter}... Skipping...')
        
```
