```python
# Import dependencies
import pandas as pd
import io
import requests
from config import gkey
import json
```

```python
# Get route data from Open Flights endpoint
url="https://raw.githubusercontent.com/jpatokal/openflights/master/data/routes.dat"
request_url = requests.get(url).content
all_routes = pd.read_csv(io.StringIO(request_url.decode('utf-8')), header=None)
```

```python
# Reset index
all_routes.reset_index(inplace=True)
```

```python
# Check DF
all_routes.head()
```

```python
# Rename clumns
all_routes.columns = ['index', 'airline', 'airlineId', 'srcIata', 'srcAirportId', 'dstIata', 'dstAirport', 'codeshare', 'stops', 'equipment']
```
